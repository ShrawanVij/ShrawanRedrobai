```python
import re

def normalize_skills(raw_skills, skill_aliases):
    normalized_skills = []
    for skill in raw_skills.split(','):
        skill = skill.strip()
        for key, aliases in skill_aliases.items():
            if skill in aliases or skill == key:
                normalized_skills.append(key)
                break
        else:
            normalized_skills.append(skill)
    return normalized_skills

raw_skills = "Pyhton, MachineLearning, SQL, pandas, numpy, Deep-learning, JavaScrpit, Reacts, Node.JS, MongoDb, REST api, HTML/CSS, Java, Spring Boot, MySql, Microservices, Docker, kubernates, Python, TensorFlow, Keras, NLP, BERT, data-viz, matplotlib, C++, Algoritms, Data Structure, competitive programming, python, javascript, vue.js, python, flask, PostgreSQL, AWS, CI/CD, Python, Sklearn, XGboost, feature engineering, SQL, tableau, Java, Android, Kotlin, Firebase, REST, UI/UX, figma, Reactjs, TypeScrpit, GraphQL, redux, tailwind, nodejs, jest, python, R, statistics, ML, regression, clustering, Power-BI"

normalized_skills = normalize_skills(raw_skills, SKILL_ALIASES)
print(normalized_skills)
```

```python
def deduplicateskills(skills_list_to_deduplicate):
    return list(set(skills_list_to_deduplicate))

normalizedskills = ['python', 'deeplearning', 'pandas', 'data_visualization', 'python']
deduplicatedskills = deduplicateskills(normalizedskills)
print(deduplicatedskills)  # Output: ['python', 'deeplearning', 'pandas', 'data_visualization']
```

```python
def construct_vocabulary(resumes):
    vocabulary = set()
    for resume in resumes:
        skills = resume['skills']
        normalizedskills = normalize_skills(skills, SKILL_ALIASES)
        deduplicatedskills = deduplicateskills(normalizedskills)
        vocabulary.update(deduplicatedskills)
    return sorted(list(vocabulary))
resumes = [
    {'skills': "Pyhton, MachineLearning, SQL, pandas, numpy, Deep-learning"},
    {'skills': "JavaScrpit, Reacts, Node.JS, MongoDb, REST api, HTML/CSS"},
]
vocabulary = construct_vocabulary(resumes)
print(vocabulary)
```

```python
# Function to compute TF-IDF vectors
def compute_tfidf(resumes, vocabulary):
    tfidf_vectors = []
    for resume in resumes:
        skills = resume['skills']
        normalizedskills = normalize_skills(skills, SKILL_ALIASES)
        deduplicatedskills = deduplicateskills(normalizedskills)
        tfidf_vector = [0] * len(vocabulary)
        for i, skill in enumerate(vocabulary):
            if skill in deduplicatedskills:
                tfidf_vector[i] = 1
        tfidf_vectors.append(tfidf_vector)
    return tfidf_vectors

# Example usage
resumes = [
    {'skills': "Pyhton, Deep-learning, pandas, matplotlib"},
    {'skills': "Python, Machine Learning, NumPy, SciPy"},
]
vocabulary = construct_vocabulary(resumes)
tfidf_vectors = compute_tfidf(resumes, vocabulary)
print(tfidf_vectors)
```

```python

def constructjdvectors(job_descriptions, vocabulary):
    jd_vectors = []
    for jobdescription in job_descriptions:
        skills = jobdescription['skills']
        normalizedskills = normalize_skills(skills, SKILL_ALIASES)
        deduplicatedskills = deduplicateskills(normalizedskills)
        jd_vector = [0] * len(vocabulary)
        for i, skill in enumerate(vocabulary):
            if skill in deduplicatedskills:
                jd_vector[i] = 1
        jd_vectors.append(jd_vector)
    return jd_vectors


job_descriptions = [
    {'skills': "Python, Deep Learning, Pandas, Matplotlib"},
    {'skills': "Machine Learning, NumPy, SciPy, Data Visualization"},

]
vocabulary = construct_vocabulary(resumes)
jdvectors = constructjdvectors(job_descriptions, vocabulary)
print(jdvectors)
```

```python
import math

def computecosinesimilarity(vector1, vector2):
    dot_product = sum(x * y for x, y in zip(vector1, vector2))
    magnitude1 = math.sqrt(sum(x**2 for x in vector1))
    magnitude2 = math.sqrt(sum(x**2 for x in vector2))
    return dot_product / (magnitude1 * magnitude2)

# Function to rank resumes for each job description
def rankresumes(resumes, jdvectors, vocabulary):
    rankings = []
    for i, jdvector in enumerate(jdvectors):
        resume_scores = []
        for j, resume in enumerate(resumes):
            tfidfvector = compute_tfidf([resume], vocabulary)[0]
            score = computecosinesimilarity(tfidfvector, jdvector)
            resume_scores.append((j, score))
        resume_scores.sort(key=lambda x: x[1], reverse=True)
        rankings.append(resume_scores[:3])
    return rankings

# Example usage
resumes = [
    {'skills': "Pyhton, Deep-learning, pandas, matplotlib"},
    {'skills': "Python, Machine Learning, NumPy, SciPy"},
]
job_descriptions = [
    {'skills': "Python, Deep Learning, Pandas, Matplotlib"},
    {'skills': "Machine Learning, NumPy, SciPy, Data Visualization"},
]
vocabulary = construct_vocabulary(resumes)
jdvectors = constructjdvectors(job_descriptions, vocabulary)
rankings = rankresumes(resumes, jdvectors, vocabulary)
print(rankings)
```
